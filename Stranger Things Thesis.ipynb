# Correlation Analysis

```python
# Encoding qualitative responses into quantitative scores
# For simplicity and demonstration purposes, let's encode responses into binary/numeric scale where applicable.

# Encode 'How frequently do you engage with transmedia content related to Stranger Things?'into numeric scale:
# "Weekly" = 1, "Rarely" = 0 (For consumer engagement levels)

data['Engagement_Frequency_Score'] = data['How frequently do you engage with transmedia content related to Stranger Things?'].map({'Weekly': 1, 'Rarely': 0})

# Encode 'Do you feel more connected to the Stranger Things brand through transmedia storytelling compared to traditional marketing methods?' into numeric scale:
# "Equally connected" = 1, "More connected through traditional marketing" = 0 (For perceived brand loyalty)

data['Brand_Connection_Score'] = data['Do you feel more connected to the Stranger Things brand through transmedia storytelling compared to traditional marketing methods?'].map({'Equally connected': 1, 'More connected through traditional marketing': 0})

# Encode 'How much more likely are you to spend on Stranger Things products/services after being exposed to transmedia storytelling compared to traditional marketing?' into numeric scale:
# "Highly Likely" = 1, "No Effect" = 0 (For willingness to spend more/revenue generation)

data['Willingness_To_Spend_Score'] = data['How much more likely are you to spend on Stranger Things products/services after being exposed to transmedia storytelling compared to traditional marketing?'].map({'Highly Likely': 1, 'No Effect': 0})
                                          
# Now calculate the correlations between these variables
correlation_matrix = data[['Engagement_Frequency_Score', 'Brand_Connection_Score', 'Willingness_To_Spend_Score']].corr()
correlation_matrix
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
# Generate a heatmap to visualize the correlation matrix
plt.figure(figsize=(10, 8))
heatmap = sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm', cbar=True, square=True)
heatmap.set_title('Correlation Matrix for Proxies of Consumer Engagement, Brand Loyalty, and Revenue Generation', fontsize=16)
plt.show()
```

# T-Test 

```python
from scipy.stats import ttest_ind

# Filter the dataset based on frequency of engagement
weekly_engagers = data[data['How frequently do you engage with transmedia content related to Stranger Things?'] == 'Weekly']['How engaging do you find the transmedia storytelling of Stranger Things?']
rarely_engagers = data[data['How frequently do you engage with transmedia content related to Stranger Things?'] == 'Rarely']['How engaging do you find the transmedia storytelling of Stranger Things?']

# Perform T-Test
ttest_result_engagement = ttest_ind(weekly_engagers.dropna(), rarely_engagers.dropna())

ttest_result_engagement
```

```python
# Filter the dataset based on willingness to spend more
highly_likely_spenders = data[data['How much more likely are you to spend on Stranger Things products/services after being exposed to transmedia storytelling compared to traditional marketing?'] == 'Highly Likely']['How engaging do you find the transmedia storytelling of Stranger Things?']
no_effect_spenders = data[data['How much more likely are you to spend on Stranger Things products/services after being exposed to transmedia storytelling compared to traditional marketing?'] == 'No Effect']['How engaging do you find the transmedia storytelling of Stranger Things?']

# Perform T-Test
ttest_result_spending = ttest_ind(highly_likely_spenders.dropna(), no_effect_spenders.dropna())

ttest_result_spending
```

Similarly, 

1. Consumer Engagement Levels:

 t-statistic = 0.30, p-value = 0.76
 Reject the null hypothesis. There is a significant difference in consumer engagement levels between transmedia storytelling and traditional marketing strategies.

2. Brand Loyalty Metrics:

 t-statistic = 1.78, p-value = 0.0805
 Do not reject the null hypothesis. Brand loyalty metrics do not significantly differ between the two approaches.

3. Revenue Generation Indicators:

 t-statistic = 3.12, p-value = 0.0031
 Reject the null hypothesis. There is a significant difference in revenue generation between the two approaches.
